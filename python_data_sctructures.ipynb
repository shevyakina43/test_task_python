25. –°—Ç–≤–æ—Ä–∏ –¥–≤—ñ –º–Ω–æ–∂–∏–Ω–∏:

`A = {1, 2, 3, 4}`  
`B = {3, 4, 5, 6}`


—ñ –∑–Ω–∞–π–¥–∏:  
- –æ–±‚Äô—î–¥–Ω–∞–Ω–Ω—è (union)
- –ø–µ—Ä–µ—Ç–∏–Ω (intersection)
- —Ä—ñ–∑–Ω–∏—Ü—é (difference)

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

union_result = A | B
intersection_result = A & B
difference_result = A - B

print("Union:", union_result)
print("Intersection:", intersection_result)
print("Difference:", difference_result)
```

26. –ü–µ—Ä–µ–≤—ñ—Ä, —á–∏ `A` —î –ø—ñ–¥–º–Ω–æ–∂–∏–Ω–æ—é `B`

```python
is_subset = A.issubset(B)
print(is_subset)
```

## 4. Dictionaries (–°–ª–æ–≤–Ω–∏–∫–∏) ‚Äî 10 –∑–∞–≤–¥–∞–Ω—å

31. –°—Ç–≤–æ—Ä–∏ —Å–ª–æ–≤–Ω–∏–∫ `person` –∑ –∫–ª—é—á–∞–º–∏: `"name"`, `"age"`, `"city"`.

```python
person = {"name": "Alice", "age": 30, "city": "Kyiv"}
print(person)
```

32. –í–∏–≤–µ–¥–∏ –∑–Ω–∞—á–µ–Ω–Ω—è –∫–ª—é—á–∞ `'name'`

```python
print(person["name"])
```

35. –í–∏–¥–∞–ª–∏ –∫–ª—é—á `'city'`

```python
person["age"] = 31
print(person)
```

```python
print("name" in person)
```

37. –ü—Ä–æ–π–¥–∏ —Ü–∏–∫–ª–æ–º –ø–æ —Å–ª–æ–≤–Ω–∏–∫—É —ñ –≤–∏–≤–µ–¥–∏ –≤—Å—ñ –∫–ª—é—á—ñ —Ç–∞ –∑–Ω–∞—á–µ–Ω–Ω—è.

```python
for key, value in person.items():
    print(f"{key}: {value}")
```

38. –°—Ç–≤–æ—Ä–∏ —Å–ª–æ–≤–Ω–∏–∫ –æ—Ü—ñ–Ω–æ–∫ —Å—Ç—É–¥–µ–Ω—Ç—ñ–≤:  
`grades = {"Anna": 90, "Bohdan": 82, "Olena": 95}`  
—ñ –≤–∏–≤–µ–¥–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π –±–∞–ª

```python
grades = {"Anna": 90, "Bohdan": 82, "Olena": 95}
average_grade = sum(grades.values()) / len(grades)
print(average_grade)
```

39. –û–±‚Äô—î–¥–Ω–∞–π –¥–≤–∞ —Å–ª–æ–≤–Ω–∏–∫–∏ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, `dict1 | dict2` –∞–±–æ —á–µ—Ä–µ–∑ `update()`)

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}

combined_dict = dict1 | dict2
print(combined_dict)

combined_dict2 = dict1.copy()
combined_dict2.update(dict2)
print(combined_dict2)
```
