````python
# Розв'язання для завдань 36–40
# 36
import pandas as pd

customers = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4],
    'Name': ['Anna', 'Maria', 'Taras', 'Mark'],
    'City': ['Kyiv', 'Lviv', 'Odesa', 'Kyiv']
})

orders = pd.DataFrame({
    'OrderID': [101, 102, 103],
    'CustomerID': [1, 2, 1],
    'Amount': [250, 400, 150]
})

merged_df = pd.merge(customers, orders, on='CustomerID')
merged_df


# 37
new_orders = pd.DataFrame({
    'OrderID': [104, 105],
    'CustomerID': [3, 4],
    'Amount': [300, 500]
})

all_orders = pd.concat([orders, new_orders], ignore_index=True)
all_orders

# 38
left_join = pd.merge(customers, orders, on='CustomerID', how='left')

no_orders = left_join[left_join['Amount'].isna()]
no_orders

# 39
left_join.to_csv('customers_orders.csv', index=False)

df_from_csv = pd.read_csv('customers_orders.csv')
df_from_csv

# 40  
left_join.dtypes
df_from_csv.dtypes


# merge → соединять таблицы,  concat - добавлять строки,  left join + NaN - находить отсутствие данных,  to_csv / read_csv - сохранять и загружать данные,  dtypes - проверять типы данных
````

````python
# Розв'язання для завдань 61–70
# 61
import pandas as pd

df = pd.read_csv("customers_orders.csv")
df
````

````python
df['Date'] = pd.to_datetime([
    '2024-01-05', '2024-01-15', '2024-02-10',
    '2024-02-20', '2024-03-05'
])

df['Product'] = ['Apple', 'Apple', 'Milk', 'Bread', 'Apple']
df['Category'] = ['Fruit', 'Fruit', 'Dairy', 'Bakery', 'Fruit']
df['Price'] = [2.5, 2.5, 3.0, 2.0, 2.5]
df['Quantity'] = [100, 60, 80, 50, 0]

df
````

````python
df['Revenue'] = df['Price'] * df['Quantity']
df
````

````python
# 62
df['Month'] = df['Date'].dt.to_period('M')

monthly_revenue = df.groupby('Month')['Revenue'].sum()

monthly_revenue.plot(kind='line', marker='o')
plt.title('Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
````

````python
# 63
top_products = (
    df.groupby('Product')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(3)
)

top_products.plot(kind='bar')
plt.title('Top 3 Products by Revenue')
plt.ylabel('Revenue')
plt.show()
````

````python
# 64
city_revenue = df.groupby('City')['Revenue'].sum()

city_revenue.plot(kind='bar')
plt.title('Revenue by City')
plt.ylabel('Revenue')
plt.show()
````

````python
# 65
corr = df[['Amount', 'Price', 'Quantity', 'Revenue']].corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Heatmap')
plt.show()
````

````python
# 66
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

monthly_revenue.plot(ax=axes[0], title='Revenue by Month')

df.groupby('Category')['Revenue'].sum().plot(
    kind='bar', ax=axes[1], title='Revenue by Category'
)

plt.tight_layout()
plt.show()
````

````python
# 67
df['Year'] = df['Date'].dt.year

yearly = df.groupby(['Year', 'Month'])['Revenue'].sum().unstack(0)

yearly.plot()
plt.title('Revenue by Month (2 Years)')
plt.ylabel('Revenue')
plt.show()
````

````python
# 68
import matplotlib.pyplot as plt

city_revenue = df.groupby('City')["Revenue"].sum()

fig, ax1 = plt.subplots()

city_revenue.plot(kind='bar', ax=ax1)
ax1.set_ylabel('Revenue')

ax2 = ax1.twinx()
df.groupby('City')['Price'].mean().plot(
    color='black', ax=ax2, marker='o'
)
ax2.set_ylabel('Average Price')

plt.title('Revenue and Avg Price by City')
plt.show()
````
