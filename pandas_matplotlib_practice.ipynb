````python
# Розв'язання для завдань 21–26
# Підказка: використовуй np.nan для створення пропусків.
import pandas as pd
import numpy as np
# 21
df = pd.DataFrame({
    'Name':['Anna','Maria', None,'Mark','Max'],
    'Age':[15, np.nan, 65, None, 43],
    'City': ['Kyiv', None,'New York','Paris', np.nan],
    'Salary': [50000, 60000, None, 55000, np.nan]
})
# 22
text_columns = df.select_dtypes(include='object').columns
df[text_columns] = df[text_columns].fillna('Unknown')   # select - выбрать,  - dtypes - типы данных, include='object - включить текстовые, .columns - взять только имена столбцов, df[text_columns] - обратиться к этим столбцам, .fillna('Unknown') - заменить NaN

# 23
num_columns = df.select_dtypes(include='number').columns
df[num_columns] = df[num_columns].fillna(df[num_columns].mean())

# 24
df_dropped = df.dropna() # Удали строки, где есть NaN

# 25
df.isna().sum()

# 26
df.dtypes
df.head()  
# Твій код тут ↓
````
